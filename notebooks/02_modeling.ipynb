Model training

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

```

```python
df = pd.read_csv("../data/raw/boston.csv")
x = df.drop(columns=['MEDV'],axis=1)
y = df['MEDV']
```

Feature Scaling = normalization + standardization

```python
scalar = StandardScaler()
x_scaled = scalar.fit_transform(x)
```

Train-Test Split

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x_scaled,y,train_size=0.2,random_state=42)
```

linear regression model

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
y_pred = lr.predict(x_test)
```

Evaluate using Cost Function

```python
from sklearn.metrics import root_mean_squared_error

rmse_linear = root_mean_squared_error(y_test, y_pred)
rmse_linear
```

Regularization

Ridge Regression (L2)

```python
from sklearn.linear_model import Ridge

ridge = Ridge(alpha=1.0)
ridge.fit(x_train, y_train)
rmse_ridge= root_mean_squared_error(y_test, ridge.predict(x_test))
```

Lasso Regression (L1)

```python
from sklearn.linear_model import Lasso
lasso = Lasso(alpha=1.0)
lasso.fit(x_train, y_train)
rmse_lasso = root_mean_squared_error(y_test, lasso.predict(x_test))
```

Compare Results

```python
print("Linear RMSE:", rmse_linear)
print("Ridge RMSE:", rmse_ridge)
print("Lasso RMSE:", rmse_lasso)
```
